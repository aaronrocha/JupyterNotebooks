Determine the slope of the line L tangent to the parabola y = x^2 at the point P(a, a^2)

We cannot find the slope with only one point so we begin with a second line whose slpe we can compute, a secant line.

Our secant line passes thru P and Q(b, b^2)
let us write h = b - a, for the difference  of the x-coordinates of P and Q.
Then the coordinates of Q are given by the formulas
b = a + h and b^2 = (a + h)^2
hence the difference in the y-coordinates of P and Q is 
b^2 - a^2 = (a + h)^2 - a^2
