Find the derivatives

```python
p1 = 4 * x - 5
p1
```

```python
diff(p1)
```

```python
plot(p1, (x,-5,5))
```

```python
p2 = -16*t**2 + 100
p2
```

```python
diff(p2)
```

```python
g1 = plot(p2, show=False)
g2 = plot(Out[7], show=False)
g1.append(g2[0])
g1.show()
```

Notice that the graph of the derivative is positive when it is increasing and negative when its decreasing.  Remember the derivative is the slope of the tangent line.  Also note when the graph of the derivative touches the x-axis, when the slope is zero.
